# Python matplotlib
_________________
* It is a visualization utility
* It is a open source graph plotting library

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10])
y=np.array([0,100])
plt.plot(x,y)
plt.show()
```

Plot function

It draws a line from one point to another point.
It has two parameters first paramater is x axis points and second parameter is y axis points

```python
x=np.array([0,10,20])
y=np.array([0,100,200])
plt.plot(x,y,'o')
```

```python
x=np.array([2,3,8,10])
y=np.array([4,9,3,12])
plt.plot(x,y)
y=np.array([2,3,8,10])
plt.plot(y)
```

```python
plt.plot(y,marker='o')
```

```python
plt.plot(y,marker='*')
plt.plot(y,marker='D')
```

Syntax: Marker|line style|colour Line style
'-' Solid line ':' Dotted line '--' Dashed line '-.' Dashed/dotted line

'r' Red 'g' Green 'b' Blue 'c' Cyan 'm' Magenta 'y' Yellow 'k' Black 'w' White
plt.plot(y,'D-g')

```python
plt.plot(y,'D:g')
plt.plot(y,'D--r')
```

```python
plt.plot(y,'D-.b')
plt.plot(y,'D-.y',ms=10)
```

```python
plt.plot(y,'o-.g',ms=10,mec='b',mfc='r')
plt.plot(y,'o-.g',ms=10,mec='#ff00ff',mfc='#ffff00')
```

```python
plt.plot(y,'o-.g',ms=10,mec='blue',mfc='pink')
plt.plot(y,linestyle='dashed')
```

```python
plt.plot(y,ls=':')
plt.plot(y,ls='')
```

```python
plt.plot(y,linestyle='dashed',c='b')
```

```python
plt.plot(y,linestyle='dashed',c='b',lw=10)
```

```python
y1=np.array([2,3,8,10])
y2=np.array([1,4,9,12])
plt.plot(y1,linestyle='-',c='b')
plt.plot(y2,linestyle='-',c='r')
```

```python
x1=np.array([2,4,6,8])
x2=np.array([2,4,6,8])
y1=np.array([2,3,8,10])
y2=np.array([1,4,9,12])
plt.plot(x1,y1,x2,y2)
```

```python
x1=np.array([1,2,3,4,5])
x2=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
y2=np.array([10,8,12,12,15])
plt.plot(x1,y1,x2,y2)
plt.xlabel('overs')
plt.ylabel('runs')
plt.title('ODI Ind vs Aus')
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('ODI Ind vs Aus',fontdict=font1,loc='left')#font and title location
plt.xlabel('overs',fontdict=font2)
plt.ylabel('runs',fontdict=font2)
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
plt.title('India')
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
plt.scatter(x1,y1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
plt.scatter(x2,y2)
plt.title('Ind vs Aus')
plt.show()
```

```python
plt.scatter(x1,y1,c='r')
plt.scatter(x2,y2,c='g')
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1)
x2=np.array([1,2,3,4,5])
y2=np.array([10,7,16,19,2])
c2=np.array(['cyan','black','grey','brown','pink'])
plt.scatter(x2,y2,c=c2)
plt.title('Ind vs Aus')
plt.show()
```

```python
x1=np.array([1,2,3,4,5])
y1=np.array([14,18,8,10,9])
c1=np.array(['red','green','blue','orange','magenta'])
plt.scatter(x1,y1,c=c1,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x1,y1,c=c1,cmap='Accent')
```

```python
np.random.seed(0)
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors)

plt.show()
```

```python
x = np.random.randint(10, size=(10))
y = np.random.randint(10, size=(10))
colors = np.random.randint(10, size=(10))
size=np.array([10,15,20,50,100,30,40,25,55,60])
plt.scatter(x, y, c=colors,s=size)
```

```python
plt.scatter(x, y, c=colors,s=size,alpha=0.5)
```

```python
import numpy as np
a=np.random.normal(170,10,250)
plt.hist(a)
plt.show()
```

```python
a=np.array([40,25,15,20])
plt.pie(a)
plt.show()
```

```python
a=np.array([40,25,15,20])
l=np.array(["Ind","Aus","Eng","Pak"])
plt.pie(a,labels=l)
plt.show()
```

```python
plt.pie(a,labels=l,startangle=180)
```

```python
ex=[0,0,0,0.2]
plt.pie(a,labels=l,startangle=180,explode=ex)
```

```python
c=["black","r","#00ff00","pink"]
ex=[0.2,0.2,0.2,0.2]
plt.pie(a,labels=l,startangle=180,explode=ex,colors=c)
```

```python
a=np.array([40,25,15,20])
l=["Ind","Aus","Eng","Pak"]
c=["b","r","#00ff00","pink"]
plt.pie(a,labels=l,startangle=180,colors=c)
plt.legend(title="Scores")
plt.show()
```
